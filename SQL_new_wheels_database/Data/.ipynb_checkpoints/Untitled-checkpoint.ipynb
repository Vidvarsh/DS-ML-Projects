```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named q8
# Sample data
# q8 = pd.DataFrame({
#     'Qtr': [1, 2, 3, 4],
#     'Number_of_Orders': [100, 120, 90, 110],
#     'Qtr_revenue': [20000, 22000, 18000, 21000]
# })

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Number_of_Orders on the first Y-axis (left)
ax1.plot(q8['Qtr'], q8['Number_of_Orders'], color='tab:blue', marker='o', label='Number of Orders')
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Number of Orders', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second Y-axis (right)
ax2 = ax1.twinx()

# Plot Qtr_revenue on the second Y-axis (right)
ax2.plot(q8['Qtr'], q8['Qtr_revenue'], color='tab:red', marker='s', label='Qtr Revenue')
ax2.set_ylabel('Qtr Revenue', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Set the title
plt.title('Dual-Axis Line Chart: Number of Orders vs. Qtr Revenue')

# Show the plot
plt.show()
```
