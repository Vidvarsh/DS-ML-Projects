### Outlier Treatment

```python
for i in election[num]:
    plt.figure(figsize=(5,2))
    sns.set_style('whitegrid')
    plt.boxplot(x=election[i],vert=False,showmeans=True)
    plt.xlabel(i)
    plt.show()
```

Observations:
- From the boxplots above, 'economic.cond.national' and 'economic.cond.household' have outliers

```python
def outlier_percentage(x):
    col_ol_counter=0
    min_val=x.min()
    max_val=x.max()
    Q3=np.percentile(x,75)
    Q1=np.percentile(x,25)
    IQR=Q3-Q1
    ll=Q1-(1.5*IQR)
    ul= Q3+(1.5*IQR)
    for value in x:
        if value<ll or value>ul:
            col_ol_counter+=1
    return round(((col_ol_counter/1517)*100),2)
```

```python
election_num=election[num]
```

```python
for i in election_num.columns:
    percent=outlier_percentage(election[i])
    print('The outlier percent of {} is {}%'.format(i,percent))
```

Observations:
- The outliers present int he said columns are very minimal.
- However, in order to understand their impact on the model performance, they can be treated. In this case, we can use the IQR method.

```python
def treat_outlier(x):
    Q3=np.percentile(x,75)
    Q1=np.percentile(x,25)
    IQR=Q3-Q1
    ll=Q1-(1.5*IQR)
    ul= Q3+(1.5*IQR)
    return x.apply(lambda y:ll if y<ll else y).apply(lambda y:ul if y>ul else y)
```

```python
election_ol_treated=pd.DataFrame(election.copy())
```

```python
for i in election_ol_treated[num]:
    election_ol_treated[i]=treat_outlier(election_ol_treated[i])
```

```python
election_ol_treated.describe().T
```

```python
for i in election_ol_treated[num]:
    print('Skewness of {}: {}'.format(i,round(skew(election_ol_treated[i]),2)))
```

Observations:
- We can observe that after the outlier treatment, the skewness of the treated columns have reduced.

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
```

```python
logr_model=LogisticRegression()
logr_model.fit(X_train,y_train)
y_predict_logr=logr_model.predict(X_test)
y_predict_logr_train=logr_model.predict(X_train)
```

```python
print('Train Data Performance')
print('\nLogistic Regression model score:\n',np.round(logr_model.score(X_train,y_train),2))
print('\nConfusion Matrix:\n',confusion_matrix(y_train,y_predict_logr_train))
print('\nClassification report:\n',classification_report(y_train,y_predict_logr_train))
```

```python
#cm plt
cm_logr_train=confusion_matrix(y_train, y_predict_logr_train)

df_cm_logr_train = pd.DataFrame(cm_logr_train, index = [i for i in ["Conservative","Labour"]],
                  columns = [i for i in ["Conservative","Labour"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm_logr_train, annot=True ,fmt='g',cmap='crest')
plt.title('Confusion Matrix- Logistic Regression- Train Data')
```

```python
#roc curve
probs_logr_train=logr_model.predict_proba(X_train)
probs_logr_train=probs_logr_train[:,1]
train_fpr_logr, train_tpr_logr, train_thresholds_logr=roc_curve(y_train,probs_logr_train)
plt.plot([0,1],[0,1],linestyle='--',color='r')
plt.plot(train_fpr_logr,train_tpr_logr)
plt.title('Logistic Regression- Train Data- ROC Curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
```

### KNN model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
k_values=np.arange(1,21,1)
cv_scores=[]
for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn,X_train,y_train,cv=5,scoring='accuracy')
    cv_scores.append(scores.mean())
plt.plot(k_values,cv_scores,marker='o',color='green')
plt.title('K values vs Cross Validation Accuracy Scores to find optimal k')
plt.xlabel('k Values')
plt.ylabel('Cross Validation accuracy scores')
plt.axvline(x=9,linestyle='--',color='r')
plt.axhline(y=max(cv_scores),linestyle='--',color='r');
```

```python
optimal_k=k_values[cv_scores.index(max(cv_scores))]
print('Optimal k value: ',optimal_k)
```

```python
k_values=np.arange(1,21,1)
cv_scores_f1=[]
for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn,X_train,y_train,cv=5,scoring='f1')
    cv_scores_f1.append(scores.mean())
plt.plot(k_values,cv_scores_f1,marker='o')
plt.title('K values vs Cross Validation F1 Scores to find optimal k')
plt.xlabel('k Values')
plt.ylabel('Cross Validation f1 scores')
plt.axvline(x=9,linestyle='--',color='r')
plt.axhline(y=max(cv_scores_f1),linestyle='--',color='r');
```

```python
optimal_k_f1=k_values[cv_scores_f1.index(max(cv_scores_f1))]
print('Optimal k value: ',optimal_k_f1)
```

## Find the number of characters, words and sentences for the mentioned documents. (Hint: use .words(), .raw(), .sent() for extracting counts)

```python
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids()
nltk.download('punkt')
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
speeches=[
    ('Roosevelt','1941-Roosevelt.txt'),
    ('Kennedy','1961-Kennedy.txt'),
    ('Nixon','1973-Nixon.txt')
]
```

```python
for president,fileid in speeches:
    speech_content=inaugural.raw(fileid)
    speech_words=inaugural.words(fileid)
    speech_sentences=inaugural.sents(fileid)
    no_of_chars=len(speech_content)
    no_of_words=len(speech_words)
    no_of_sentences=len(speech_sentences)
    print("\n{}'s speech:".format(president))
    print('\nNumber of characters:{}' .format(no_of_chars))
    print('Number of words:{}' .format(no_of_words))
    print('Number of Sentences:{}' .format(no_of_sentences))
```

## Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.

```python
import string
```

```python
for president,fileid in speeches:
    speech_content=inaugural.raw(fileid)
    speech_words=inaugural.words(fileid)
    speech_sentences=inaugural.sents(fileid)
    no_of_chars=len(speech_content)
    no_of_words=len(speech_words)
    no_of_sentences=len(speech_sentences)
    print("\n",president,"'s speech: ")
    print('Number of words before removal of stopwords:',no_of_words)
    stopwords=nltk.corpus.stopwords.words('english')+list(string.punctuation)+list(['--'])
    clean_words=[]
    for word in inaugural.words(fileid):
        word=word.lower()
        if word not in stopwords:
            clean_words.append(word)
    no_of_clean_words=len(clean_words)
    print('Number of words after removal of stopwords and punctuations:',no_of_clean_words)
    sample_sentence=" ".join(clean_words[:20])
    print('Sample sentence: ', sample_sentence)
```

## Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords)

```python
from collections import Counter
from nltk.tokenize import word_tokenize
```

```python
for president,fileid in speeches:
    speech_content=inaugural.raw(fileid)
    speech_words=inaugural.words(fileid)
    speech_sentences=inaugural.sents(fileid)
    no_of_chars=len(speech_content)
    no_of_words=len(speech_words)
    no_of_sentences=len(speech_sentences)
    print("\n",president,"'s speech: ")
    stopwords=nltk.corpus.stopwords.words('english')+list(string.punctuation)+list(['--'])
    clean_words=[]
    for word in inaugural.words(fileid):
        word=word.lower()
        if word not in stopwords:
            clean_words.append(word)
    no_of_clean_words=len(clean_words)
    word_count=Counter(clean_words)
    most_common_words=word_count.most_common(3)
    print('Most common words:',most_common_words)
```
