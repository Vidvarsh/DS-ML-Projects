## Problem 2


An independent research organization is trying to estimate the probability that an accident at a nuclear power plant will result in radiation leakage. The types of accidents possible at the plant are, fire hazards, mechanical failure, or human error. The research organization also knows that two or more types of accidents cannot occur simultaneously.

According to the studies carried out by the organization, the probability of a radiation leak in case of a fire is 20%, the probability of a radiation leak in case of a mechanical 50%, and the probability of a radiation leak in case of a human error is 10%. The studies also showed the following;

The probability of a radiation leak occurring simultaneously with a fire is 0.1%.
The probability of a radiation leak occurring simultaneously with a mechanical failure is 0.15%.
The probability of a radiation leak occurring simultaneously with a human error is 0.12%.
On the basis of the information available, answer the questions below:


### 2.1 What are the probabilities of a fire, a mechanical failure, and a human error respectively?


```python
pr_r_given_f=0.2
pr_r_given_m=0.5
pr_r_given_h=0.1
pr_r_and_f= 0.001
pr_r_and_m= 0.0015
pr_r_and_h= 0.0012
```

```python
pr_f=pr_r_and_f/pr_r_given_f
pr_m=pr_r_and_m/pr_r_given_m
pr_h=pr_r_and_h/pr_r_given_h
print('The probailities of a fire, a mechanical failure, and a human error are ', round(pr_f,3),',',round(pr_m,3),',',round(pr_h,3),'respectively')
```

### 2.2 What is the probability of a radiation leak?

```python
pr_r=((pr_r_given_f*pr_f)+(pr_r_given_m*pr_m)+(pr_r_given_h*pr_h))
print('The probability of a radiation leak is ', round(pr_r,4))
```

### 2.3 Suppose there has been a radiation leak in the reactor for which the definite cause is not known. What is the probability that it has been caused by:

- A Fire.
- A Mechanical Failure.
- A Human Error.

```python
pr_f_given_r= (pr_r_given_f*pr_f)/pr_r
pr_m_given_r= (pr_r_given_m*pr_m)/pr_r
pr_h_given_r= (pr_r_given_h*pr_h)/pr_r
print('The probability of the radiation leak caused by fire is ', round(pr_f_given_r,3))
print('The probability of the radiation leak caused by mechanical failure is ', round(pr_m_given_r,3))
print('The probability of the radiation leak caused by human error is ', round(pr_h_given_r,3))
```

### 3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
mu=5
sigma=1.5
```

```python
from scipy.stats import norm
x1=np.linspace(0,10,100)
plt.plot(x1,norm.pdf(x1,mu,sigma))
plt.title('Breaking Strength of Gummy Bags')
plt.xlabel('Breaking strength')
plt.ylabel('Probability')
plt.axvline(3.17,color='r')
x2=np.linspace(0,3.17,100)
plt.fill_between(x2,norm.pdf(x2,mu,sigma),color='r')
```

## Problem 4:

Grades of the final examination in a training course are found to be normally distributed, with a mean of 77 and a standard deviation of 8.5. Based on the given information answer the questions below.

### 4.1 What is the probability that a randomly chosen student gets a grade below 85 on this exam?

```python
mu=77
sigma=8.5
```

```python
x=np.linspace(50,110,100)
plt.plot(x,norm.pdf(x,mu,sigma))
plt.title('Grades of Final Exam')
plt.xlabel('Marks')
plt.ylabel('Probability')
plt.axvline(85,color='green')
x2=np.linspace(50,85,100)
plt.fill_between(x2,norm.pdf(x2,mu,sigma),color='green')
```

### 4.2 What is the probability that a randomly selected student scores between 65 and 87?

```python
x=np.linspace(50,110,100)
plt.plot(x,norm.pdf(x,mu,sigma))
plt.title('Grades of Final Exam')
plt.xlabel('Marks')
plt.ylabel('Probability')
plt.axvline(65,color='pink')
plt.axvline(87,color='pink')
x2=np.linspace(65,87,100)
plt.fill_between(x2,norm.pdf(x2,mu,sigma),color='pink')
```

### 4.3 What should be the passing cut-off so that 75% of the students clear the exam?

```python
x=np.linspace(50,110,100)
plt.plot(x,norm.pdf(x,mu,sigma))
plt.title('Grades of Final Exam')
plt.xlabel('Marks')
plt.ylabel('Probability')
cut_off=stats.norm.ppf(0.25,mu,sigma)
plt.axvline(cut_off,color='yellow')
x1=np.linspace(50,cut_off,100)
plt.fill_between(x1,stats.norm.pdf(x1,mu,sigma),color='red')
x2=np.linspace(cut_off,110,100)
plt.fill_between(x2,stats.norm.pdf(x2,mu,sigma),color='green')
plt.text(cut_off-15,0.04,'Fail',color='red',fontsize=15)
plt.text(cut_off+20,0.04,'Pass',color='green',fontsize=15)
```

```python
print('The passing cut-off so that 75% of the students clear the exam is', round(cut_off,2))
```

## Problem 5:

Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);

```python
zingaro=pd.read_csv('Zingaro_Company.csv')
```

```python
zingaro.head()
```

```python
zingaro.info()
```

```python
round(zingaro.describe().T,2)
```

### 5.1 Earlier experience of Zingaro with this particular client is favorable as the stone surface was found to be of adequate hardness. However, Zingaro has reason to believe now that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

```python
from scipy.stats import ttest_1samp
```

One tail, one sample t test.

Establishing Hypothesis:
- Null hypothesis: The unpolished stones are suitable for printing
        H0: Hardness>=150
- Alternate Hypothesis: The unpolished stones are not suitable for printing
        H1: Hardness<150
Alpha:0.05

```python
hardness=150
t_stat,p_val=ttest_1samp(zingaro['Unpolished '],hardness)
```

```python
p_value=p_val/2
```

```python
alpha=0.05
```

```python
if p_value<alpha:
    print('Null hypothesis rejected. Considering a confidence of 95%, the unpolished stones are not suitable for printing')
else:
    print('Failed to reject null hypothesis. The unpolished stones are suitable for printing')
```

### 5.2 Is the mean hardness of the polished and unpolished stones the same?

Independent two sample t test.

Establishing the hypothesis:
- Null Hypothesis: The mean hardness of the polished and unpolished stones are the same
      H0: mu_pol = mu_unpol
- Alternate Hypothesis: The mean hardness of the polished and unpolished stones are not the same
       H1: mu_pol not = mu_unpol
       
Alpha = 0.05

```python
from scipy.stats import ttest_ind
```

```python
t_statistic, p = ttest_ind(zingaro['Treated and Polished'],zingaro['Unpolished '])
```

```python
if p<alpha:
    print('Null hypothesis rejected.Considering a confidence of 95%, the mean hardness of polished and unpolished stones are not equal.')
else:
    print('Failed to Reject Null. The mean hardness of polished and unpolished stones are equal')
```

## Problem 6:

Aquarius health club, one of the largest and most popular cross-fit gyms in the country has been advertising a rigorous program for body conditioning. The program is considered successful if the candidate is able to do more than 5 push-ups, as compared to when he/she enrolled in the program. Using the sample data provided can you conclude whether the program is successful? (Consider the level of Significance as 5%)

Note that this is a problem of the paired-t-test. Since the claim is that the training will make a difference of more than 5, the null and alternative hypotheses must be formed accordingly.

```python
gym=pd.read_csv('Aquarius_gym.csv')
```

```python
gym.head()
```

```python
gym.info()
```

```python
round(gym.describe().T,2)
```

One tail, Paired t test

Establishing hypothesis:

- Null Hypothesis: The difference betwen the mean number of sit-ups done by the candidates after and before the training is less than 5
        H0: mu_after-mu_before < 5
- Alternative hypothesis: The difference betwen the mean number of sit-ups done by the candidates after and before the training is atleast 5
        H1: H0: mu_after-mu_before >= 5
        
Alpha=0.05

```python
from scipy.stats import ttest_rel
t_stats, p_val= stats.ttest_rel(gym['Before'],gym['After'])
p_value=p_val/2
```

```python
if p_value< alpha:
    print('Null hypothesis rejected. At 95% confidence, the difference between the mean number of sit-ups done by the candidates after and before the training is atleast 5. Hence, the program is successful.')
else:
    print('Failed to reject null hypothesis. At 95% confidence, the difference between the mean number of sit-ups done by the candidates after and before the training is less than 5. Hence, the program is not successful.')
    
```

To test for normality, we carry out Shapiro and Anderson test for normality


Alloy 1:
- H0: The response distribution of responses is normal
- H1: The response distribution of responses is not normal

```python
# Shapiro test for normality
t1,p1=stats.shapiro(d1d1)
t2,p2=stats.shapiro(d1d2)
t3,p3=stats.shapiro(d1d3)
t4,p4=stats.shapiro(d1d4)
t5,p5=stats.shapiro(d1d5)

t6,p6=stats.shapiro(d1m1)
t7,p7=stats.shapiro(d1m1)
t8,p8=stats.shapiro(d1m3)

t9,p9=stats.shapiro(d1t1)
t10,p10=stats.shapiro(d1t2)
t11,p11=stats.shapiro(d1t3)


print(round(p1,8))
print(round(p2,8))
print(round(p3,8))
print(round(p4,8))
print(round(p5,8))
print(round(p6,10))
print(round(p7,10))
print(round(p8,11))
print(round(p9,11))
print(round(p10,11))
print(round(p11,11))
```

```python
print('At 95% CI, we reject the null hyposis. Hence, the response distribution for alloy 1 is not normal across different groups')
```

Since Shapiro test for normality failed, we perform the Anderson Darling test for normality

```python
#Anderson Darling test for normality
print(stats.anderson(d1d1['Response'],dist='norm'))
print(stats.anderson(d1d2['Response'],dist='norm'))
print(stats.anderson(d1d3['Response'],dist='norm'))
print(stats.anderson(d1d4['Response'],dist='norm'))
print(stats.anderson(d1d5['Response'],dist='norm'))

print(stats.anderson(d1m1['Response'],dist='norm'))
print(stats.anderson(d1m2['Response'],dist='norm'))
print(stats.anderson(d1m3['Response'],dist='norm'))

print(stats.anderson(d1t1['Response'],dist='norm'))
print(stats.anderson(d1t2['Response'],dist='norm'))
print(stats.anderson(d1t3['Response'],dist='norm'))
```

From the Anderson- Darling test result, if the test statistic is less than the critical value for a significance of 5%, we can fail to reject normal, i.e., the distribution is normal. Thus, for the given groups using Alloy 1, the following are the results

Hence, the ANOVA assumptions of normality fails for the above groups.

- For the method category, with or without interaction the p value is less than 0.05.
